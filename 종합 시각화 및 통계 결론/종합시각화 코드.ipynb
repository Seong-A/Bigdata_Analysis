```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#    
df = pd.read_excel('./data/separate_data.xlsx', sheet_name=None)

#    
font_path = '/Library/Fonts/Arial Unicode.ttf'  

#  
fontprop = fm.FontProperties(fname=font_path).get_name()
plt.rcParams["font.family"] = fontprop

#   
df_region1_reordered = df['  1']
df_region2_reordered = df['  2']

#   
labels1 = df_region1_reordered['']
percentages_emergency1 = df_region1_reordered['']
percentages_in1 = df_region1_reordered[' ()']
percentages_out1 = df_region1_reordered[' ()']
percentages_work1 = df_region1_reordered['']
percentages_young1 = df_region1_reordered[' ']
percentages_old1 = df_region1_reordered[' ']

labels2 = df_region2_reordered['']
percentages_emergency2 = df_region2_reordered['']
percentages_in2 = df_region2_reordered[' ()']
percentages_out2 = df_region2_reordered[' ()']
percentages_work2 = df_region2_reordered['']
percentages_young2 = df_region2_reordered[' ']
percentages_old2 = df_region2_reordered[' ']

width = 0.15
x1 = np.arange(len(labels1))
x2 = np.arange(len(labels2))

fig, axs = plt.subplots(2, 1, figsize=(12, 10))
axs[0].bar(x1 - width*2, percentages_emergency1, width, label='', color='b')
axs[0].bar(x1 - width, percentages_in1, width, label=' ()', color='g')
axs[0].bar(x1, percentages_out1, width, label=' ()', color='r')
axs[0].bar(x1 + width, percentages_work1, width, label='', color='m')
axs[0].bar(x1 + width*2, percentages_young1, width, label=' ', color='y')
axs[0].bar(x1 + width*3, percentages_old1, width, label=' ', color='c')

axs[1].bar(x2 - width*2, percentages_emergency2, width, label='', color='b')
axs[1].bar(x2 - width, percentages_in2, width, label=' ()', color='g')
axs[1].bar(x2, percentages_out2, width, label=' ()', color='r')
axs[1].bar(x2 + width, percentages_work2, width, label='', color='m')
axs[1].bar(x2 + width*2, percentages_young2, width, label=' ', color='y')
axs[1].bar(x2 + width*3, percentages_old2, width, label=' ', color='c')

axs[0].set_title(' ', fontproperties=fontprop, fontsize=16)
axs[1].set_xlabel('', fontproperties=fontprop, fontsize=14)
axs[0].set_ylabel('', fontproperties=fontprop, fontsize=14)
axs[1].set_ylabel('', fontproperties=fontprop, fontsize=14)

axs[0].set_xticks(x1)
axs[0].set_xticklabels(labels1, rotation=45, fontsize=12)
axs[1].set_xticks(x2)
axs[1].set_xticklabels(labels2, rotation=45, fontsize=12)

axs[0].legend(loc='upper right')
axs[1].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
