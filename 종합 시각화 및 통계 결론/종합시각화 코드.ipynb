```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 엑셀 파일에서 데이터 읽기
df = pd.read_excel('./data/separate_data.xlsx', sheet_name=None)

# 한글 폰트 경로 설정
font_path = '/Library/Fonts/Arial Unicode.ttf'  

# 폰트 설정
fontprop = fm.FontProperties(fname=font_path).get_name()
plt.rcParams["font.family"] = fontprop

# 시트별 데이터 추출
df_region1_reordered = df['지역별 데이터 1']
df_region2_reordered = df['지역별 데이터 2']

# 막대 그래프 그리기
labels1 = df_region1_reordered['지역별']
percentages_emergency1 = df_region1_reordered['응급의료기관']
percentages_in1 = df_region1_reordered['총전입 (명)']
percentages_out1 = df_region1_reordered['총전출 (명)']
percentages_work1 = df_region1_reordered['일자리']
percentages_young1 = df_region1_reordered['저출산 백분율']
percentages_old1 = df_region1_reordered['고령화 백분율']

labels2 = df_region2_reordered['지역별']
percentages_emergency2 = df_region2_reordered['응급의료기관']
percentages_in2 = df_region2_reordered['총전입 (명)']
percentages_out2 = df_region2_reordered['총전출 (명)']
percentages_work2 = df_region2_reordered['일자리']
percentages_young2 = df_region2_reordered['저출산 백분율']
percentages_old2 = df_region2_reordered['고령화 백분율']

width = 0.15
x1 = np.arange(len(labels1))
x2 = np.arange(len(labels2))

fig, axs = plt.subplots(2, 1, figsize=(12, 10))
axs[0].bar(x1 - width*2, percentages_emergency1, width, label='응급의료기관', color='b')
axs[0].bar(x1 - width, percentages_in1, width, label='총전입 (명)', color='g')
axs[0].bar(x1, percentages_out1, width, label='총전출 (명)', color='r')
axs[0].bar(x1 + width, percentages_work1, width, label='일자리', color='m')
axs[0].bar(x1 + width*2, percentages_young1, width, label='저출산 백분율', color='y')
axs[0].bar(x1 + width*3, percentages_old1, width, label='고령화 백분율', color='c')

axs[1].bar(x2 - width*2, percentages_emergency2, width, label='응급의료기관', color='b')
axs[1].bar(x2 - width, percentages_in2, width, label='총전입 (명)', color='g')
axs[1].bar(x2, percentages_out2, width, label='총전출 (명)', color='r')
axs[1].bar(x2 + width, percentages_work2, width, label='일자리', color='m')
axs[1].bar(x2 + width*2, percentages_young2, width, label='저출산 백분율', color='y')
axs[1].bar(x2 + width*3, percentages_old2, width, label='고령화 백분율', color='c')

axs[0].set_title('제목을 모로하딩', fontproperties=fontprop, fontsize=16)
axs[1].set_xlabel('지역별', fontproperties=fontprop, fontsize=14)
axs[0].set_ylabel('백분율', fontproperties=fontprop, fontsize=14)
axs[1].set_ylabel('백분율', fontproperties=fontprop, fontsize=14)

axs[0].set_xticks(x1)
axs[0].set_xticklabels(labels1, rotation=45, fontsize=12)
axs[1].set_xticks(x2)
axs[1].set_xticklabels(labels2, rotation=45, fontsize=12)

axs[0].legend(loc='upper right')
axs[1].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
