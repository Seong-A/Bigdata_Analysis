```python
#라이브러리 불러오기
import pandas as pd 

#파일경로를 찾고 변수 file_path에 저장
file_path= './data/people_why_seoul.xlsx'

#read_excel()함수로 데이터프레임 변환 
df = pd.read_excel(file_path)


#데이터프레임 출력
df
```

```python
import numpy
from scipy import stats

#귀무가설1: 인구수는 고령화 수가 높으면 높다.
#대립가설1: 인구수는 고령화 수가 높으면 낮다.
#귀무가설2: 인구수는 출산 수가 높으면 높다.
#대립가설2: 인구수는 출산 수가 낮으면 낮다.

y = df['계'].to_list()
ys_young = df['저출산'].to_list()         
ys_old = df['고령화'].to_list()
y_in=df['총전입'].to_list()   
y_out=df['총전출'].to_list()   
y_medical=df['응급의료기관'].to_list()   
y_job=df['일자리'].to_list()  

# people
people = numpy.array(y)
# young
young = numpy.array(ys_young)
# old
old = numpy.array(ys_old)
#in
people_in = numpy.array(y_in)
#out
people_out = numpy.array(y_out)
#medical
medical = numpy.array(y_medical)
#job
job = numpy.array(y_job)

correlation_coef1, young_value = stats.pearsonr(young, people)
print('상관계수 = %.5f, young_pvalue = %.20f' % (correlation_coef1, young_value))
correlation_coef2, old_value = stats.pearsonr(old, people)
print('상관계수 = %.5f, old_pvalue = %.20f' % (correlation_coef2, old_value))
correlation_coef3, people_in_value =stats.pearsonr(people_in, people)
print('상관계수 = %.5f, p-value = %.20f' % (correlation_coef3, people_in_value))
correlation_coef4, people_out_value = stats.pearsonr(people_out, people)
print('상관계수 = %.5f, p-value = %.20f' % (correlation_coef4, people_out_value))
correlation_coef5, medical_value = stats.pearsonr(medical, people)
print('상관계수 = %.5f, p-value = %.20f' % (correlation_coef5, medical_value))
correlation_coef6, job_value = stats.pearsonr(job, people)
print('상관계수 = %.5f, p-value = %.20f' % (correlation_coef6, job_value))
```
