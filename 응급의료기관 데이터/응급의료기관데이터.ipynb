```python
#라이브러리 불러오기
import pandas as pd 

#파일경로를 찾고 변수 file_path에 저장
file_path= './data/hospital_.xlsx'

#read_excel()함수로 데이터프레임 변환 
df3 = pd.read_excel(file_path)

#데이터프레임 출력
df3
```

```python
#백분율로 표현시 데이터 차이가 너무 나서 백분율 삭제
df3 = df3.drop(labels='백분율',axis=1)
df3 = df3.drop(14)
df3
```

```python
####응급의료기관 수로 표현

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 경로 설정
font_path = '/Library/Fonts/Arial Unicode.ttf'  

# 폰트 설정
fontprop = fm.FontProperties(fname=font_path)

labels = df3['지역별']
hospitals = df3['계']

plt.figure(figsize=(10, 6))

plt.bar(range(len(labels)), hospitals, color='hotpink')

plt.title('지역별 응급의료기관 수', fontproperties=fontprop) 
plt.xlabel('지역별', fontproperties=fontprop)  
plt.ylabel('수', fontproperties=fontprop) 

plt.xticks(range(len(labels)), labels, rotation='0', fontproperties=fontprop)  # x 축 레이블 지정

plt.show()
```
