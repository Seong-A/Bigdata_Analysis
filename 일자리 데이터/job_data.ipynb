```python
#라이브러리 불러오기
import pandas as pd

#파일경로 찾고 변수 file_path에 저장
file_path = './data1/job_data.xlsx'

#read_excel()함수로 데이터프레임 변환
df = pd.read_excel(file_path)

#출력
df
```

```python
df_area1 = df[df['시도별'].isin(['서울특별시','경기도','부산광역시','대구광역시','인천광역시','광주광역시','대전광역시','울산광역시','세종특별자치시'])]

df_area2 = df[df['시도별'].isin(['강원도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주도','계'])]


data1 = ['서울특별시','경기도','부산광역시','대구광역시','인천광역시','광주광역시','대전광역시','울산광역시','세종특별자치시']
df_area1_re = df_area1.set_index('시도별').reindex(data1).reset_index()

data2 = ['강원도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주도','계']
df_area2_re = df_area2.set_index('시도별').reindex(data2).reset_index()

print("시도별 데이터 1 : \n",df_area1_re)
print("시도별 데이터 2 : \n",df_area2_re)
```

```python
import matplotlib.pylab as plt

plt.figure(figsize=(15,20))

plt.subplot(2,1,1)
plt.xlabel('시도별')
plt.ylabel('일자리')
plt.bar(df_area1_re['시도별'],df_area1_re['백분율'],width=0.5,color='red')

plt.subplot(2,1,2)
plt.xlabel('시도별')
plt.ylabel('일자리')
plt.bar(df_area2_re['시도별'],df_area2_re['백분율'],width=0.5,color='blue')
```
