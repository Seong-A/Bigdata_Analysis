```python
import matplotlib.pylab as plt
import numpy

#한글 깨짐 현상으로 인한 폰트 적용
plt.rcParams['font.family'] = 'Malgun Gothic'

xs=df.index.to_list()         
ys_young=df['저출산'].to_list()         
ys_old=df['고령화'].to_list()

plt.figure(figsize=(20, 10))

a=numpy.arange(len(xs))

# 저출산은 빨강, 고령화는 파랑으로 색을 지정하여 막대 그래프 그리기
       


plt.bar(a-0.0, ys_young, label="저출산", width=0.3, color='red')  
plt.bar(a+0.3, ys_old, label="고령화", width=0.3, color='blue')
plt.xticks(a,xs)

plt.legend()
plt.xlabel('지역명')
plt.ylabel('인구수')
plt.ylim(10000,3300000)
plt.title('저출산 고령화 그래프')
```

```python

from scipy import stats  
from pandas import DataFrame
from scipy.stats import pearsonr

#귀무가설1: 인구수와 고령화 수는 관계가 없다
#대립가설1: 인구수와 고령화 수는 관계가 있다.
#귀무가설2: 인구수와 출산 수는 관계가 없다
#대립가설2: 인구수와 출산 수는 관계가 있다.

y=df['계'].to_list()
ys_young=df['저출산'].to_list()         
ys_old=df['고령화'].to_list()

#people
people = numpy.array(y)
#young
young = numpy.array(ys_young)
#old
old = numpy.array(ys_old)

correlation_coef1, young_value = pearsonr(young, people)
print('상관계수 = %.5f, young_pvalue = %.20f'%(correlation_coef1, young_value))
correlation_coef1, old_value = pearsonr(young, old)
print('상관계수 = %.5f, old_pvalue = %.20f'%(correlation_coef1, old_value))
```
